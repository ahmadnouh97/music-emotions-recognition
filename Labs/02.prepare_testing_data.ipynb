```python
data_dir = os.path.join('Data', 'test_data')
happy_dir = os.path.join(data_dir, 'happy')
happy_features_dir = os.path.join(happy_dir, 'features')
happy_features_path = os.path.join(happy_features_dir, 'features.csv')
```

```python
happy_data = pd.read_csv(happy_features_path, index_col=False)
```

```python
happy_data.rename(columns={'music_id': 'SongTitle', 'label': 'Label', 'lyrics': 'Lyrics'}, inplace=True)
```

```python
happy_data['songID'] = happy_data['SongTitle'].apply(lambda title: int(title.split('.')[0]))
```

```python
happy_data['SongTitle'] = happy_data['SongTitle'].apply(lambda title: title.split('.')[1])
```

```python
data_dir = os.path.join('Data', 'test_data')
sad_dir = os.path.join(data_dir, 'sad')
sad_features_dir = os.path.join(sad_dir, 'features')
sad_features_path = os.path.join(sad_features_dir, 'features.csv')
```

```python
sad_data = pd.read_csv(sad_features_path, index_col=False)
```

```python
sad_data.rename(columns={'music_id': 'SongTitle', 'label': 'Label', 'lyrics': 'Lyrics'}, inplace=True)
```

```python
sad_data['songID'] = sad_data['SongTitle'].apply(lambda title: int(title.split('.')[0]))
```

```python
sad_data['SongTitle'] = sad_data['SongTitle'].apply(lambda title: title.split('.')[1])
```

```python
data = pd.concat([happy_data, sad_data], axis=0)
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

```python
data.to_csv(
    os.path.join(data_dir, 'data.csv'),
    index=None
)
```
