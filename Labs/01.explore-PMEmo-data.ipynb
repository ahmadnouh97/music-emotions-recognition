```python
from pydub import AudioSegment
```

```python
songs_path = Path('../Data/PMEmo2019/songs')
songs_wav_path = Path('../Data/PMEmo2019/songs_wav')
```

```python
songs_wav_path.mkdir(parents=True, exist_ok=True)
```

```python
filenames = os.listdir(str(songs_path))
```

```python
AudioSegment.converter = "ffmpeg"
```

```python
for name in filenames:
    src = str(songs_path / name)
    dist = str(songs_wav_path / name[:-4]) + '.wav'
    print(f'src = {src}')
    print(f'dist = {dist}')
    sound = AudioSegment.from_mp3(src)
    sound.export(dist , format='wav')
```
